```python
import networkx as nx
import matplotlib.pyplot as plt
#from networkx.algorithms import community

# Загрузим датасет
url = "facebook_combined.txt"
G = nx.read_edgelist(url)
with open(url, 'r') as f:
    lines = f.readlines()

# Берем только половину данных
half_lines = lines[:len(lines)//2]

# Создаем граф из половины данных
G = nx.parse_edgelist(half_lines, nodetype=int)

```

```python
print(G)
```

```python
# Обнаружение сообществ с использованием алгоритма Лувена
import community

communities = community.best_partition(G)
```

```python
pos = nx.spring_layout(G)
colors = [communities[node] for node in G.nodes()]
```

```python
plt.figure(figsize=(10, 8))
nx.draw(G, pos, node_color=colors, with_labels=False, cmap=plt.cm.RdYlBu, node_size=30)
plt.title("Community Detection in Facebook Social Circles")
plt.show()
```
